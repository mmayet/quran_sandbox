```python
#All imports
import json
from pyarabic import araby
```

```python
# removes arabic vowels and stylistic spacing
def strip_text(text):
    return araby.strip_tatweel(araby.strip_tashkeel(text.replace('۞','')))
```

```python
#Read quran.json
f = open('data/quran.json', 'r', encoding="utf-8")
data = json.load(f)
```

```python
#View contents of data for Sūrah al-Fātiḥah
data[0]
```

```python
#Test string clean up
print(data[0]['verses'][0]['text'])
print(strip_text(data[0]['verses'][0]['text']))
```

```python
#Construct and view a list of each word of the sūrah
entire_surah = ''
for i in data[0]['verses']:
    entire_surah += strip_text(i['text'])
    if data[0]['total_verses'] != i['id']:
        entire_surah += ' '
entire_surah = entire_surah.split(' ')
entire_surah
```

```python
#Build and add list of each word of the sūrah
for surah in data:
    entire_surah_text = ''
    for i in surah['verses']:
        entire_surah_text += strip_text(i['text'])
        if surah['total_verses'] != i['id']:
            entire_surah_text += ' '
    entire_surah_list = entire_surah_text.split(' ')
    data[surah['id']-1]['entire_surah_list'] = list(set(entire_surah_list))
    #print(f"Wrote entire_surah_list for Sūrah #{surah['id']}: {surah['name']}")
```

```python
def get_all_other_surahs_text_list(id):
    all_other_surahs_text_list = []
    for surah in data:
        if surah['id'] != id:
            all_other_surahs_text_list.extend(surah['entire_surah_list'])
    return all_other_surahs_text_list
```

### Below are the results of the words that are unique to each respect _sūrah_ in _theory_.
